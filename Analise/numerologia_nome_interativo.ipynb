# Analise Numerologica de Nomes (Interativo)

Este notebook executa a analise por etapas e permite escolher recursos complementares com resposta `sim/nao` ou `s/n` (maiusculas/minusculas).

```python
# Etapa 1: Tabela de correspondencia numerologica
TABELA_NUMEROLOGICA = {
    'a': 1, 'j': 1, 's': 1,
    'b': 2, 'k': 2, 't': 2,
    'c': 3, 'l': 3, 'u': 3,
    'd': 4, 'm': 4, 'v': 4,
    'e': 5, 'n': 5, 'x': 5,
    'f': 6, 'o': 6, 'y': 6,
    'g': 7, 'p': 7, 'z': 7,
    'h': 8, 'q': 8,
    'i': 9, 'r': 9,
}

TABELA_NUMEROLOGICA
```

```python
# Etapa 2: Funcoes auxiliares
import unicodedata

def remover_acentos(texto: str) -> str:
    normalizado = unicodedata.normalize('NFD', texto)
    return ''.join(ch for ch in normalizado if unicodedata.category(ch) != 'Mn')

def perguntar_sim_nao(pergunta: str) -> bool:
    while True:
        resp = input(f"{pergunta} [sim/nao ou s/n]: ").strip().lower()
        if resp in {'sim', 's'}:
            return True
        if resp in {'nao', 'n'}:
            return False
        print('Resposta invalida. Use sim/nao ou s/n.')
```

```python
# Etapa 3: Analise numerologica
def analisar_nome(nome: str, tratar_acentos: bool = True):
    texto = remover_acentos(nome) if tratar_acentos else nome

    soma = 0
    detalhes = []
    ignoradas = []

    for ch in texto.lower():
        if not ch.isalpha():
            continue

        if ch in TABELA_NUMEROLOGICA:
            valor = TABELA_NUMEROLOGICA[ch]
            soma += valor
            detalhes.append((ch, valor))
        else:
            ignoradas.append(ch)

    return soma, detalhes, ignoradas

def reduzir_numero(valor: int):
    passos = [valor]
    while valor > 9:
        valor = sum(int(d) for d in str(valor))
        passos.append(valor)
    return valor, passos
```

```python
# Etapa 4: Fluxo interativo com opcoes complementares
print('=== Analise Numerologica de Nome ===')
nome = input('Digite um nome proprio: ').strip()

if not nome:
    print('Nenhum nome foi informado.')
else:
    usar_tratamento_acentos = perguntar_sim_nao('Deseja tratar acentos automaticamente?')
    calcular_reducao = perguntar_sim_nao('Deseja calcular o numero reduzido final?')
    analisar_partes = perguntar_sim_nao('Deseja analisar cada parte do nome separadamente?')

    soma, detalhes, ignoradas = analisar_nome(nome, tratar_acentos=usar_tratamento_acentos)

    if not detalhes:
        print('Nao foi possivel calcular: nenhuma letra da tabela foi encontrada.')
    else:
        expressao = ' + '.join(str(v) for _, v in detalhes)
        pares = ', '.join(f"{l}:{v}" for l, v in detalhes)

        print(f'\nNome informado: {nome}')
        print(f'Correspondencias usadas: {pares}')
        print(f'Calculo: {expressao} = {soma}')

        if ignoradas:
            letras_ignoradas = ', '.join(sorted(set(ignoradas)))
            print(f'Letras sem valor na tabela (ignoradas): {letras_ignoradas}')

        if calcular_reducao:
            reduzido, passos = reduzir_numero(soma)
            cadeia = ' -> '.join(str(p) for p in passos)
            print(f'Reducao numerologica: {cadeia}')
            print(f'Numero final reduzido: {reduzido}')

        if analisar_partes:
            print('\nAnalise por partes do nome:')
            for parte in [p for p in nome.split() if p.strip()]:
                parte_soma, parte_detalhes, _ = analisar_nome(parte, tratar_acentos=usar_tratamento_acentos)
                if parte_detalhes:
                    print(f'- {parte}: {parte_soma}')
                else:
                    print(f'- {parte}: sem letras da tabela')
```
