```python
vendas = pd.read_csv('01.amazon_sales_dataset.csv')
vendas.columns
```

```python
vendas.info()
```

```python
coluna_alvo = 'total_sales'
# Medidas de Tendências Centrais
total_vendas = vendas[coluna_alvo].sum()
media_valor_vendas = vendas[coluna_alvo].mean()
venda_minima = vendas[coluna_alvo].min()
venda_maxima = vendas[coluna_alvo].max()
```

```python
# Medidas de Posição
q1 = np.percentile(vendas[coluna_alvo], 25)
q2 = np.percentile(vendas[coluna_alvo], 50)
# q2 Mediana
q3 = np.percentile(vendas[coluna_alvo], 75)
```

```python
# Avaliar a confiabilidade da média
delta_media_mediana = media_valor_vendas - q2
distancia_percentual = media_valor_vendas / q2
# Caso 1: Distancia <= 10% -> (0.9 ~ 1.1) Média é altamente confiável
# Caso 2: Distancia entre 10% e 25% -> (0.75 ~0.9 / 1.1 ~ 1.25) Média é moderadamente confiável
# Caso 3: Distancia > 25% -> (0.75 ~ 1.25) Média é pouco confiável, ela deve ser distorcida nos valores extremos, ou seja, outliers.
```

```python
# Assimetria (Skewness)
assimetria = vendas[coluna_alvo].skew()
# Caso 1: Skewness entre -0.5 e 0.5 -> Distribuição aproximadamente simétrica R = equilibrada
# Caso 2: Skewness > 0.5 (Positiva) R = Valores altos são raros, porém são tão grandes que distorcem o cálculo
# Caso 3: Skewness < -0.5 (Negativa) R = Valores baixos são raros, porém são tão grandes que distorcem o cálculo
```

```python
print ('\n --- Resumo Executivo de Vendas ---')
print(f'1. VolumeTotal de Vendas: (Soma das Vendas): ${total_vendas:,.2f}')
# O formato :,.2f é usado para formatar o número com vírgulas como separadores de milhar e duas casas decimais.
# (story telling): Representa o tamanho total da nossa operação no período analisado, indicando a escala de vendas alcançada.
print(f'2. Gasto Médio: $ {media_valor_vendas:,.2f}')
# (story telling): Valor esperado que um clientye gaste em média conosco.
# Média de Valor das Vendas: {media_valor_vendas}')
print(f'3. Maior Venda: ${venda_maxima:,.2f}')
# (story telling): Nosso Record. investigar o perfil do cliente 
# O valor máximo de venda registrado, indicando a maior transação realizada.
print(f'4. Menor Venda: ${venda_minima:,.2f}')
# (story telling): Investigar se há anomalias ou fraude# O valor mínimo de venda registrado, indicando a menor transação realizada.
```

```python
print('--- Relatório Estatístico de Negócio ---')
print(f'A Média é $ {media_valor_vendas:,.2f}') 
print(f'A mediana (O Valor Central) é $ {q2:,.2f}')
print(f' O Delta (A Diferença Financeira entre elas) é $ {delta_media_mediana:,.2f}')
print('\n --- Análise Comportamental ---')
print(f'Distancia Relativa entre Média e Mediana: {distancia_percentual:.2f}')
if abs(distancia_percentual-1) <= 0.1:
    print("INFERÊNCIA: Baixa Dispersão. A média é altamente confiável para representar a coluna.")
elif abs(distancia_percentual-1) <= 0.25:
    print("INFERÊNCIA: Dispersão Moderada. Atenção a valores extremos.")
else:
    print("INFERÊNCIA: Alta Dispersão. A média é não confiável para representar a coluna, considere utilizar a mediana.")

print(f"Grau de Assimetria (Skewness): {assimetria:.4f}")
if assimetria > 0.5:
    print("INFERÊNCIA: (Assimétrica positiva): cauda longa à direita.")
    # Story telling: Os grandes clientes são minoria, todavia faturam valores expressivos, ou seja, são os responsáveis por grande parte do faturamento.
elif assimetria < -0.5:
    print("INFERÊNCIA: (Assimétrica negativa): cauda longa à esquerda.")
    # Story telling: O valor médio está sendo puxado para baixo por um grande número de vendas de baixo valor, talvez liquidação.
else:
    print("INFERÊNCIA: (Simétrica): distribuição equilibrada. Média e mediana estão próximas, indicando uma distribuição equilibrada dos valores de vendas.")
 
```

```python
# Navegação e Filtros Avançados

# iloc: Acessar por posição -> pega 'n' linhas por índice
# usabilidade:

amostra = vendas.iloc[0:10] # pega as 10 primeiras linhas
print(amostra)
```

```python
# loc: Acessar por rótulo -> pega 'n' linhas por rótulo (usa o nome da coluna e filtra com base em alguma condição)
# quem são os melhores clientes? (top 10 clientes que mais gastaram)
top_clientes = vendas.loc[vendas['total_sales'] > q3] # filtra as vendas acima de 75% do valor das vendas
top_clientes2 = vendas.loc[vendas['total_sales'] > q3, ['order_id', 'customer_name', 'category', 'total_sales']] # filtra as vendas acima de 75% do valor das vendas e seleciona apenas as colunas 'order_id', 'customer_name', 'category' e 'total_sales'
print(top_clientes2)
```

```python
# query: Acessar por consulta -> pega 'n' linhas por consulta (usa uma string para filtrar os dados com base em alguma condição)
# O cliente deseja saber quais pedidos da categoria eletrõnico ficaram acima da média geral
filtro = f"category == 'Electronics' and total_sales > {media_valor_vendas}"
eletronicos = vendas.query(filtro)
print(eletronicos)
```
