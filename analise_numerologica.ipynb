```python
# Etapa 1: Tabela de correspondencia numerologica
TABELA_NUMEROLOGICA = {
    'a': 1, 'j': 1, 's': 1,
    'b': 2, 'k': 2, 't': 2,
    'c': 3, 'l': 3, 'u': 3,
    'd': 4, 'm': 4, 'v': 4,
    'e': 5, 'n': 5, 'x': 5,
    'f': 6, 'o': 6, 'y': 6,
    'g': 7, 'p': 7, 'z': 7,
    'h': 8, 'q': 8,
    'i': 9, 'r': 9,
}

TABELA_NUMEROLOGICA
```

```python
# Etapa 2: Funcoes auxiliares
import unicodedata

def remover_acentos(texto):
    normalizado = unicodedata.normalize('NFD', texto)
    return ''.join(ch for ch in normalizado if unicodedata.category(ch) != 'Mn')

def perguntar_sim_nao(pergunta):
    while True:
        resp = input(f'{pergunta} [s/n]: ').strip().lower()
        if resp in {'sim', 's'}:
            return True
        if resp in {'nao', 'n'}:
            return False
        print('Resposta invalida. Use s ou n.')
```

```python
# Etapa 3: Analise numerologica e reducao
def analisar_nome(nome, tratar_acentos=True):
    texto = remover_acentos(nome) if tratar_acentos else nome
    soma, detalhes, ignoradas = 0, [], []
    for ch in texto.lower():
        if not ch.isalpha():
            continue
        if ch in TABELA_NUMEROLOGICA:
            valor = TABELA_NUMEROLOGICA[ch]
            soma += valor
            detalhes.append((ch, valor))
        else:
            ignoradas.append(ch)
    return soma, detalhes, ignoradas

def reduzir_numero(valor):
    passos = [valor]
    while valor > 9:
        valor = sum(int(d) for d in str(valor))
        passos.append(valor)
    return valor, passos
```

```python
# Etapa 4: Fluxo interativo com opcoes complementares
print('=== Analise Numerologica de Nome ===')
nome = input('Digite um nome proprio: ').strip()

if not nome:
    print('Nenhum nome foi informado.')
else:
    usar_acentos  = perguntar_sim_nao('Tratar acentos automaticamente?')
    calc_reducao  = perguntar_sim_nao('Calcular numero reduzido final?')
    anal_partes   = perguntar_sim_nao('Analisar cada parte do nome separadamente?')

    soma, detalhes, ignoradas = analisar_nome(nome, tratar_acentos=usar_acentos)

    if not detalhes:
        print('Nao foi possivel calcular: nenhuma letra da tabela foi encontrada.')
    else:
        expressao = ' + '.join(str(v) for _, v in detalhes)
        pares     = ', '.join(f'{l}:{v}' for l, v in detalhes)
        print(f'\nNome: {nome}')
        print(f'Correspondencias: {pares}')
        print(f'Calculo: {expressao} = {soma}')

        if ignoradas:
            print(f'Ignoradas (sem valor na tabela): {chr(44).join(sorted(set(ignoradas)))}')

        if calc_reducao:
            reduzido, passos = reduzir_numero(soma)
            print(f'Reducao: {" -> ".join(str(p) for p in passos)}')
            print(f'Numero final: {reduzido}')

        if anal_partes:
            print('\nAnalise por parte do nome:')
            for parte in nome.split():
                ps, pd, _ = analisar_nome(parte, tratar_acentos=usar_acentos)
                print(f'  {parte}: {ps}' if pd else f'  {parte}: sem letras da tabela')
```
